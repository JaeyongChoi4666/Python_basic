```python
from selenium import webdriver
import bs4

url="https://www.coupang.com/np/categories/497244?listSize=60&brand=&offerCondition=&filterType=rocket_wow%2Ccoupang_global&isPriceRange=false&minPrice=&maxPrice=&page=1&channel=user&fromComponent=N&selectedPlpKeepFilter=&sorter=bestAsc&filter=&component=497144&rating=0&rocketAll=true"

options = webdriver.ChromeOptions()
options = webdriver.ChromeOptions()
options.add_argument("disable-gpu")   
options.add_argument("lang=ko_KR")    
options.add_argument('user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36')  # user-agent 


driver = webdriver.Chrome(executable_path="driver/chromedriver.exe",options=options)
driver.get(url)

response = driver.page_source
driver.close()

soup = bs4.BeautifulSoup(response, 'html.parser')

arr=[]
for tmp in soup.select("#productList > li"):
    name=tmp.select_one("a > dl > dd > div.name").text
    if tmp.select_one("a > dl > dd > div.price-area > div.price-wrap > div.price > em > strong.price-value")!=None:
        price=tmp.select_one("a > dl > dd > div.price-area > div.price-wrap > div.price > em > strong.price-value").text
    else:
        price=tmp.select_one("a > dl > dd > div.price-area > div > strong:nth-child(4)").text
    print(name[5:-1])
    print(price)
    data={"name":name[5:-1],"price":price}
    arr.append(data)
    
print("="*50)
print(arr)
```

```python
import hdfs
import json

client_hdfs = hdfs.InsecureClient('http://1.234.5.158:9870', user='hdfs')
client_hdfs.delete('/id125/ex_20210525_coupang.json')

with client_hdfs.write('/id125/ex_20210525_coupang.json', encoding="utf-8") as writer:
    json.dump(arr,writer)
```
