```python
import csv
import folium

file = open('csv/부산광역시_부산진구_가로수현황_20201123.csv','r',encoding="euc-kr")
rdr = csv.reader(file)

next(rdr,None) #메타데이터 스킵

#읽은 내용 보기
#위치명[0], 위도[1], 경도[2]
for idx,tmp in enumerate(rdr):
    if idx == 0: #최초 1번 객체 만들기, 시작위치의 중심이 location에 정의된 곳
        map_data = folium.Map(location=[float(tmp[1]),float(tmp[2])], zoom_size=12, zoom_start=13)
    
    folium.Marker([float(tmp[1]),float(tmp[2])],popup=tmp[0]).add_to(map_data)
    print(idx,tmp[0][11:],tmp[1],tmp[2])
    
map_data.save('map/ex01.html')
```

```python
import csv
import requests
import pandas as pd 

csv_url="http://ihongss.com/resources/train.csv"

with requests.Session() as s:
    download = s.get(csv_url)
    decode_content = download.content.decode('utf-8')
    file = csv.reader(decode_content.splitlines(), delimiter=",")
    
    data = list(file)
    df = pd.DataFrame(data[1:],columns=data[0])
df
```
